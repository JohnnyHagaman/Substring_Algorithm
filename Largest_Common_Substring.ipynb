```python
def largest_common_substring(str1,str2):
    """
    Takes in two strings, str1 and str2, and returns the longest common substring within both strings.
    """
    if len(str1) < len(str2):
        short = str1
        lng = str2
    else:
        short = str2
        lng = str1
        
    arr = np.zeros((len(short), len(lng))) # Create a numpy array of shape len(shorter string) x len(longer string)
    
    for i in range(len(short)): #iterate over the shorter string, rows in this instance
        for j in range(len(lng)): # iterate through longer string, columns
            
            
            if lng[j] == short[i] and (i == 0 or j == 0): # This clause specifically deals with index wrapping 
                arr[i,j] = 1                    # You should never have greater than 1 on row or column 0
                    
            elif lng[j] == short[i]:        # Else if the letters match
                arr[i,j] = arr[i-1,j-1] + 1 # add one to running total of consecutive matching letters at that matrix location
                

                
    ss_len = int(np.amax(arr)) # Find the largest value in the arr, this will be the length of the common substring
    
    in_row,in_col = np.unravel_index(np.argmax(arr, axis=None), arr.shape) # Find the index location of the max value in the array
    
    last_ind = in_col + 1 # Add one for slice notation, using column which relates to the longer string
    
    df = pd.DataFrame(arr) # I made this just for some visual context, you can return df to have a better idea of what's happening
    df.columns = [c for c in lng] # Assign the longer string elements the column name
    df.index = [c for c in short] # Assign the shorter string elements as index
    
    return df,lng[last_ind-ss_len:last_ind] #Slice of longer string returning the largest common substring
    
```
