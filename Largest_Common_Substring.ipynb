```python
def largest_common_substring(str1,str2):
    """
    Returns: Dataframe, string
    Takes in two strings, str1 and str2, and returns the longest common substring within both strings.
    Also returns a dataframe showing the matrix of size len(str1) x len(str2)
    """
        
    arr = np.zeros((len(str1), len(str2))) # Create a numpy array of shape len(str1) x len(str2) 
                                           # filled with '0.0'
    
    for i in range(len(str1)): #iterate over the str1, rows in this instance
        for j in range(len(str2)): # iterate through str2, columns
            
            
            if str2[j] == str1[i] and (i == 0 or j == 0): # This clause deals with index wrapping 
                arr[i,j] = 1                    # Never have greater than 1 on row or column 0
                    
            elif str2[j] == str1[i]:        # Else if the letters match
                arr[i,j] = arr[i-1,j-1] + 1 # add one to running total of consecutive matching letters
                

                
    ss_len = int(np.amax(arr))# Largest value in the arr, this is the length of the common substring
    
    in_row,in_col = np.unravel_index(np.argmax(arr, axis=None), arr.shape) 
                                             # Find the index location of the max value in the array
    
    last_ind = in_col + 1 # Add one for slice notation, using column which relates to str2
    
    df = pd.DataFrame(arr) # I made this just for some visual context
    df.columns = [c for c in str2] # Assign the str2 characters the column name
    df.index = [c for c in str1] # Assign the str1 characters as index
    
    return df,str2[last_ind-ss_len:last_ind] # Slice of str2, returns the largest common substring
    
```
